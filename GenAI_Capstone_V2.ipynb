```python
import pandas as pd
import numpy as np

original_data = pd.read_csv('02.csv')
pd.set_option('display.max_columns', None)
```

```python
train_data=original_data.copy()
test_data=original_data.copy()
```

```python
train_data.head()
```

```python
train_data.info()
```

```python
train_data.drop(columns=['No', 'UnitPrice', 'PricePerTsubo', 'Period', 'Remarks',
    'Renovation', 'FloorPlan', 'Purpose', 'Use', 'MunicipalityCode',
    'TimeToNearestStation', 'MaxTimeToNearestStation',
    'FrontageIsGreaterFlag', 'AreaIsGreaterFlag',
    'TotalFloorAreaIsGreaterFlag', 'Prefecture'], inplace=True)
```

```python
train_data.drop(train_data[train_data['Type']=='Agricultural Land'].index,inplace=True)
```

```python
train_data.info()
```

```python
train_data.drop(train_data[train_data['Region'].isna()].index,inplace=True)
```

```python
train_data.drop(train_data[train_data['DistrictName'].isna()].index,inplace=True)
```

```python
train_data.drop(train_data[train_data['NearestStation'].isna()].index,inplace=True)
```

```python
train_data.drop(train_data[train_data['MinTimeToNearestStation'].isna()].index,inplace=True)
```

```python
train_data.drop(train_data[train_data['LandShape'].isna()].index,inplace=True)
```

```python
train_data.drop(train_data[train_data['Frontage'].isna()].index,inplace=True)
train_data.drop(train_data[train_data['TotalFloorArea'].isna()].index,inplace=True)
train_data.drop(train_data[train_data['BuildingYear'].isna()].index,inplace=True)
train_data.drop(train_data[train_data['Structure'].isna()].index,inplace=True)
train_data.drop(train_data[train_data['Classification'].isna()].index,inplace=True)
```

```python
train_data.drop(train_data[train_data['Breadth'].isna()].index,inplace=True)
train_data.drop(train_data[train_data['CityPlanning'].isna()].index,inplace=True)
train_data.drop(train_data[train_data['CoverageRatio'].isna()].index,inplace=True)
train_data.drop(train_data[train_data['FloorAreaRatio'].isna()].index,inplace=True)
```

```python
train_data.head()
```

```python
num_cols = ['Frontage', 'TotalFloorArea', 'BuildingYear', 'Breadth', 'CoverageRatio', 'FloorAreaRatio', 'MinTimeToNearestStation', 'Area']
```

```python
cat_cols = ['Type','Region', 'Municipality', 'DistrictName', 'NearestStation', 'LandShape', 'Structure', 'Classification', 'CityPlanning', 'Direction']
```

```python
test_data=train_data.copy()
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

for col in cat_cols:
    if train_data[col].dtype == 'object':
        train_data[col]=le.fit_transform(train_data[col])
        test_data[col]=le.transform(test_data[col])
```

```python
train_data.head()
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

train_data[num_cols] = scaler.fit_transform(train_data[num_cols])
test_data[num_cols] = scaler.transform(test_data[num_cols])
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

x=train_data.drop('TradePrice', axis=1)
y=np.log1p(train_data['TradePrice']) # Apply log transformation to the target variable

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

model = LinearRegression()
model.fit(x_train, y_train)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
yTrainPrediction = model.predict(x_train)
yTestPrediction = model.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error

mseTrain = mean_squared_error(y_train, yTrainPrediction)
mseTest = mean_squared_error(y_test, yTestPrediction)

print("MSE on training set: ", mseTrain)
print("MSE on test set: ", mseTest)
```

```python
x_final_test = test_data.drop('TradePrice', axis=1)
final_predictions_log = model.predict(x_final_test)
final_predictions = np.expm1(final_predictions_log)
print(final_predictions)
```

```python
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np

mseTrain = mean_squared_error(y_train, yTrainPrediction)
rmseTrain = np.sqrt(mseTrain)
r2Train = r2_score(y_train, yTrainPrediction)

mseTest = mean_squared_error(y_test, yTestPrediction)
rmseTest = np.sqrt(mseTest)
r2Test = r2_score(y_test, yTestPrediction)

print("--- Training Set Metrics ---")
print(f"MSE on training set: {mseTrain:.4f}")
print(f"RMSE on training set: {rmseTrain:.4f}")
print(f"R-squared on training set: {r2Train:.4f}")
print("\n--- Test Set Metrics ---")
print(f"MSE on test set: {mseTest:.4f}")
print(f"RMSE on test set: {rmseTest:.4f}")
print(f"R-squared on test set: {r2Test:.4f}")
```

```python
from sklearn.ensemble import RandomForestRegressor

rf_model = RandomForestRegressor(random_state=42)
rf_model.fit(x_train, y_train)
```

```python
rf_yTrainPrediction = rf_model.predict(x_train)
rf_yTestPrediction = rf_model.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np

mseTrain_rf = mean_squared_error(y_train, rf_yTrainPrediction)
rmseTrain_rf = np.sqrt(mseTrain_rf)
r2Train_rf = r2_score(y_train, rf_yTrainPrediction)

mseTest_rf = mean_squared_error(y_test, rf_yTestPrediction)
rmseTest_rf = np.sqrt(mseTest_rf)
r2Test_rf = r2_score(y_test, rf_yTestPrediction)

print("--- RandomForestRegressor Training Set Metrics ---")
print(f"MSE on training set: {mseTrain_rf:.4f}")
print(f"RMSE on training set: {rmseTrain_rf:.4f}")
print(f"R-squared on training set: {r2Train_rf:.4f}")
print("\n--- RandomForestRegressor Test Set Metrics ---")
print(f"MSE on test set: {mseTest_rf:.4f}")
print(f"RMSE on test set: {rmseTest_rf:.4f}")
print(f"R-squared on test set: {r2Test_rf:.4f}")
```

```python
train_data['AgeOfBuilding'] = train_data['Year'] - train_data['BuildingYear']
test_data['AgeOfBuilding'] = test_data['Year'] - test_data['BuildingYear']
train_data.head()
```

```python
num_cols.append('AgeOfBuilding')
print(f"Updated numerical columns list: {num_cols}")
```

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np

scaler = MinMaxScaler()
train_data[num_cols] = scaler.fit_transform(train_data[num_cols])
test_data[num_cols] = scaler.transform(test_data[num_cols])

x = train_data.drop('TradePrice', axis=1)
y = np.log1p(train_data['TradePrice'])

x_train_new, x_test_new, y_train_new, y_test_new = train_test_split(x, y, test_size=0.2, random_state=42)

rf_model_new = RandomForestRegressor(random_state=42)

rf_model_new.fit(x_train_new, y_train_new)

rf_yTrainPrediction_new = rf_model_new.predict(x_train_new)
rf_yTestPrediction_new = rf_model_new.predict(x_test_new)

mseTrain_rf_new = mean_squared_error(y_train_new, rf_yTrainPrediction_new)
rmseTrain_rf_new = np.sqrt(mseTrain_rf_new)
r2Train_rf_new = r2_score(y_train_new, rf_yTrainPrediction_new)

mseTest_rf_new = mean_squared_error(y_test_new, rf_yTestPrediction_new)
rmseTest_rf_new = np.sqrt(mseTest_rf_new)
r2Test_rf_new = r2_score(y_test_new, rf_yTestPrediction_new)

print("\n--- RandomForestRegressor Training Set Metrics (with AgeOfBuilding) ---")
print(f"MSE on training set: {mseTrain_rf_new:.4f}")
print(f"RMSE on training set: {rmseTrain_rf_new:.4f}")
print(f"R-squared on training set: {r2Train_rf_new:.4f}")
print("\n--- RandomForestRegressor Test Set Metrics (with AgeOfBuilding) ---")
print(f"MSE on test set: {mseTest_rf_new:.4f}")
print(f"RMSE on test set: {rmseTest_rf_new:.4f}")
print(f"R-squared on test set: {r2Test_rf_new:.4f}")
```
